```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
car = pd.read_csv(r'Car_Purchasing_Data.csv',encoding='ISO-8859-1')
print(car)
```

```python
plt.scatter(car['Age'],car['Car Purchase Amount'])
plt.xlabel("Age")
plt.ylabel("Car Purchase Amount")
```

```python
plt.scatter(car['Annual Salary'],car['Car Purchase Amount'])
plt.xlabel("Annual salary")
plt.ylabel("Car Purchase Amount")
```

```python
plt.scatter(car['Credit Card Debt'],car['Car Purchase Amount'])
plt.xlabel("Credit Card Debt")
plt.ylabel("Car Purchase Amount")
```

```python
plt.scatter(car['Net Worth'],car['Car Purchase Amount'])
plt.xlabel("Net Worth")
plt.ylabel("Car Purchase Amount")
```

```python
x = car[['Age','Annual Salary','Credit Card Debt','Net Worth']]
y = car['Car Purchase Amount']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
clf = LinearRegression()
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
print(y_pred)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df1 = df.head(25)
df1
```

```python
print(clf.score(x_test, y_test))
```

```python
coeff_df = pd.DataFrame(clf.coef_, x.columns, columns=['Coefficient'])  
print(coeff_df)
```

```python
clf.intercept_
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
```

```python
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred)) 
```

```python
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
